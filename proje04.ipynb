```python
from google.colab import drive
drive.mount('/content/drive')
dosya = "/content/drive/MyDrive/Colab Notebooks/veritabani.txt"


print("Kütüphane Kataloglama Sistemime Hoş geldin :)\n")

def kitapListele():
  f = open(dosya, 'r')
  for line in f.readlines():
    print(line)
  f.close()
  menu()

def kitapKaydet():
  kitapAdi = input("Kitabın adını girin: ")
  kitapYazari = input("Kitabın yazarını girin: ")
  kitapYayinevi = input("Kitabın yayınevi bilgisini girin: ")
  kitapBasimTarihi = input("Kitabın basım tarihini girin: ")
  f = open(dosya, 'a' )
  f.write(kitapAdi + " | " + kitapYazari + " | " + kitapYayinevi + " | " + kitapBasimTarihi + " | " +'\n')
  print("Kitap kaydını başarıyla tamamladın")
  f.close()
  menu()

def kitapAra():
  kayitAra = input("Arama yapmak istediğin kitabın adını girin: ")
  kayit = open(dosya, "r")

  arananKayit = kayit.read().find(kayitAra)
  if arananKayit:
    print("Aradığın kitap kataloğumda mevcuttur.")
    menu()
  else:
    print("Aradığın kitap kataloğumda yok.")
    menu()

def cikis():
  print("\t\t\n Kataloğumdan çıkış yapmak istiyor musun?\n \t1) Evet\n  \t2) Hayır\n\t\t")
  cikisIslem = input("İşlem seç: ")
  if cikisIslem == "2":
    print("Menüye yönlendiriliyorsun. Seçimini yap. ")
    menu()
  elif cikisIslem == "1":
   print("Çıkış yapıldı.")
  else:
    print("Yanlış bir işlem yaptın. Sadece 1 ve 2 seçebilirsin. ")
    cikis()

def menu():
  print("\t\t\n Yapmak istediğin işlemi seç:\n  \t1) Kitap Listele\n  \t2) Kitap Kaydet\n  \t3) Kitap Ara\n  \t4) Çıkış Yap\n\t\t")
  islem = input("İşlem seç: ")
  if islem == "1":
    kitapListele()
  elif islem == "2":
    kitapKaydet()
  elif islem== "3":
    kitapAra()
  elif islem == "4":
    cikis()
  else:
    print("Yanlış giriş yaptın! Menüye yönlendiriliyorsun.")
    menu()

menu()






```
